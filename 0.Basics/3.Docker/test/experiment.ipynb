```python
# Connection to postgrtes via sqlaclchemy 


from sqlalchemy import create_engine 

```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
engine.connect()
```

```python
# insert table 
head = df.head(n=0)
head.to_sql(name = 'yellow_taxi_data', con=engine, if_exists='replace')
```

```python
df.head()
```

```python
df.tail(1000).to_sql(name = 'yellow_taxi_data', con=engine, if_exists='append')
```

```python
df.head(1000).to_sql(name = 'yellow_taxi_data', con=engine, if_exists='append')
```

```python
## Read values from SQL 
query = '''
SELECT * FROM  yellow_taxi_data LIMIT 10;
'''
pd.read_sql(sql=query, con=engine)
```
